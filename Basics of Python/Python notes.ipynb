```python
SA = ["a","b","d","a","c","c","b","d","a","c"]
#SA=student's mcq answers in exam
CA = ["a","d","d","b","c","c","c","d","a","c"]
#CA=correct answers of the exam
sum=0
#if 1 ans is wrong, there will be deducted an additional 0.25 i.e. total deduction=1.25 marks 
for i in range(10):
    if SA[i]==CA[i]:
        sum+=1
    else:
        sum-=0.25
print(f"Score out of 10: ",sum)
```

```python
CA = ["a","b","d","a","c","c","b","d","a","c"]
#CA=correct answers of the exam
SA = ["a","d","d","{}","c","c","c","d","a","c"]
#SA=student's mcq answers in the exam
sum=0
for i in range(10):
    if SA[i]==CA[i]:
        sum+=1
    else:
        sum-=0.25
print(f"Score out of 10: ",sum)
#programme is considering {} as a wrong answer
#before defining, programme will define any value as a wrong ans except equal value
```

```python
CA = ["a","b","d","a","c","c","b","d","a","c"]
#CA=correct answers of the exam
SA = ["a","d","d","","c","c","c","d","","c"]
#SA=student's mcq answers in the exam
sum=0
for i in range(10):
    if SA[i]==CA[i]:
        sum+=1
    elif SA[i]=="":
        sum+=0
    else:
        sum-=0.25
print(f"Score out of 10: ",sum)
#"" is defined as blank answers i.e. non-response by else if statement
```
